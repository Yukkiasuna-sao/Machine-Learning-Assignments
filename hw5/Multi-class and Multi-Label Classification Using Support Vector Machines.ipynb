```python
import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
# import warnings
# warnings.filterwarnings('ignore')

data = pd.read_csv('Anuran Calls (MFCCs)/Frogs_MFCCs.csv')
data = data.drop(['RecordID'], axis=1)

d1 = dict(zip(list(dict(data['Family'].value_counts()).keys()), list(np.arange(4))))
d2 = dict(zip(list(dict(data['Genus'].value_counts()).keys()), list(np.arange(8))))
d3 = dict(zip(list(dict(data['Species'].value_counts()).keys()), list(np.arange(10))))

for i in range(data.shape[0]):
    data.loc[i, 'Family'] = d1.get(data.loc[i, 'Family'])
    data.loc[i, 'Genus'] = d2.get(data.loc[i, 'Genus'])
    data.loc[i, 'Species'] = d3.get(data.loc[i, 'Species'])

train_data, test_data = train_test_split(data, test_size = 0.3)

train_features = train_data.iloc[:,:-3]
test_features = test_data.iloc[:,:-3]

train_total_label = train_data[['Family', 'Genus', 'Species']]
test_total_label = test_data[['Family', 'Genus', 'Species']]

data
```

# (b)
i. Accuracy score/Exact match ratio: This function calculates subset accuracy meaning the predicted set of labels should exactly match with the true set of labels.
In sklearn, metrics.accuracy_score implements this function. 

hamming score/loss methods:
Hamming Loss: The fraction of the wrong labels to the total number of labels.

## ii. Train a SVM for each of the labels using Gaussian kernels and one versus all classifiers
ii. SVM penalty: the 'C' parameter, which trades off correct classification of training examples against maximization of the decision functions margin. The width of the Gaussian Kernel: the 'gamma' parameter, which defines how far the influence of a single training example reaches, with low values meaning far and high values meaning close. 

```python
# Train a SVM for each of the labels using Gaussian kernels and one versus all classifiers
import numpy as np
# from sklearn.multiclass import OneVsRestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn.svm import SVC
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
# from sklearn.metrics import hamming_loss, accuracy_score

def ham_loss(y_true, y_pred):
    return np.sum(np.not_equal(y_true, y_pred))/float(y_true.size)

C_range = np.logspace(-1, 4, 6)
gamma_range = np.arange(0.1, 3.1, 0.1)
score = make_scorer(ham_loss, greater_is_better=False)
param_grid = dict(estimator__gamma=gamma_range, estimator__C=C_range)
# one versus rest devision function shape
svc = SVC(kernel='rbf', decision_function_shape='ovr', random_state=42)
moc = MultiOutputClassifier(svc)
grid1 = GridSearchCV(moc, param_grid=param_grid, scoring = score, cv=10)
grid1.fit(train_features, train_total_label.values)

print("The best parameters are %s with best hamming loss of %0.7f"
      % (grid1.best_params_, -grid1.best_score_))
```

## iii. L1-penalized SVMs

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

C_range = np.logspace(-1, 4, 6)
param_grid2 = dict(estimator__C=C_range)
svc2 = LinearSVC(penalty='l1', dual=False, multi_class='ovr', random_state=42)
moc2 = MultiOutputClassifier(svc2)
grid2 = GridSearchCV(moc2, param_grid=param_grid2, scoring = score, cv=10)
grid2.fit(train_features, train_total_label.values)

print("The best parameters are %s with best hamming loss of %0.7f"
      % (grid2.best_params_, -grid2.best_score_))
```

## iv. Use SMOTE to remedy class imbalance

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import KFold
C_range = np.logspace(-1, 4, 6)
cv_ham = []
cv_avg_ham = []

def cal_avg(num):
    return sum(num) / len(num)

for c in C_range:
    kf = KFold(n_splits=10)
    for train_index, test_index in kf.split(train_features):
        sm = SMOTE(random_state=42)
        x_res1, y_res1 = sm.fit_sample(train_features.iloc[train_index,:], train_data['Family'].iloc[train_index])
        x_res2, y_res2 = sm.fit_sample(train_features.iloc[train_index,:], train_data['Genus'].iloc[train_index])
        x_res3, y_res3 = sm.fit_sample(train_features.iloc[train_index,:], train_data['Species'].iloc[train_index])

        svc1 = LinearSVC(penalty='l1', dual=False, multi_class='ovr', random_state=42).fit(x_res1, y_res1)
        svc2 = LinearSVC(penalty='l1', dual=False, multi_class='ovr', random_state=42).fit(x_res2, y_res2)
        svc3 = LinearSVC(penalty='l1', dual=False, multi_class='ovr', random_state=42).fit(x_res3, y_res3)
        
        y_true1 = train_data['Family'].iloc[test_index]
        y_true2 = train_data['Genus'].iloc[test_index]
        y_true3 = train_data['Species'].iloc[test_index]
        
        y_pred1 = svc1.predict(train_features.iloc[test_index,:])
        y_pred2 = svc2.predict(train_features.iloc[test_index,:])
        y_pred3 = svc3.predict(train_features.iloc[test_index,:])
        
        y_true = pd.concat([y_true1, y_true2, y_true3], axis=1)
        y_pred = np.array([y_pred1, y_pred2, y_pred3])
        y_pred = y_pred.T

        cv_ham.append(ham_loss(y_true.values, y_pred))
        
    cv_avg_ham.append(cal_avg(cv_ham))
min_ham = min(cv_avg_ham)
best_c = C_range[cv_avg_ham.index(min_ham)]
print('The best number of penalty parameter C chosen by cross-validation is %.4f' % best_c)
print('Hamming loss:', min_ham)
```
